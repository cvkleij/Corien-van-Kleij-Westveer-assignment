```python
# looking at the data
import pandas as pd
import matplotlib.pyplot as plt
csv_data = pd.read_csv("istherecorrelation.csv", sep=";")

print(csv_data.head())

plt.plot(csv_data.iloc[:,0], csv_data.iloc[:,1])
plt.plot(csv_data.iloc[:,0], csv_data.iloc[:,2])
```

```python
# the visualisation of the data
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(csv_data.iloc[:,0], csv_data.iloc[:,1], 'g-')
ax2.plot(csv_data.iloc[:,0], csv_data.iloc[:,2], 'b-')

ax1.set_xlabel('Year')
ax1.set_ylabel('WO [x1000]', color='g')
ax2.set_ylabel('NL Beer consumption [x1000 hectoliter]', color='b')

plt.show()

#Bron: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib
```
